### Pre-annotation Sending via API for Various Annotation Tasks

In Beewant's annotation ecosystem, preannotations play a vital role across a range of tasks, streamlining and accelerating the data labeling process. Here's an overview of pre-annotation tasks for different use cases:

#### 1. Classification Pre-annotations: 
Classification tasks involve categorizing data points into predefined classes or labels. Pre-annotations in classification provide initial labels, which annotators can review and refine. Utilizing the API to send classification preannotations establishes a foundation for annotators to validate labels efficiently.


```python
# Replace the empty string with your Beewant API token
beewant_api_key = ""

# Replace the empty string with the task ID you want to associate the annotation with
task_id = ''

# Replace the empty string with the user ID who completed the annotation
user_id = ''

# Define the annotation data for classification task
annotation_data = {
    "result": [
        {
            "id": "vQ-7H0i4xn",
            "type": "choices",
            "value": {
                "choices": [
                    "Choice 1"
                ]
            },
            "origin": "manual",
            "to_name": "image",
            "from_name": "choice"
        }
    ],
    "task": task_id,
    "completed_by": user_id
}
```

```python
# Define the API endpoint URL for annotations
url = "http://beewant.com/api/tasks/TASK_ID/annotations/"

# Define the headers with the API token and content type
headers = {
    'Authorization': f'Token {beewant_api_key}',
    'Content-Type': 'application/json'
}

# Send a POST request to create annotations for classification task
response = requests.post(url, headers=headers, json=annotation_data)

# Print the response object, including status code and headers
print(response)

# Parse the JSON content from the response
data = response.json()

# Print the parsed JSON data
print(data)
```
